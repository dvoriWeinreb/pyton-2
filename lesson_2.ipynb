1.y=4*x*6

```python
x = np.linspace(0,10,100)
y = np.sin(4*x*6)
plt.plot(x, y)
plt.title('y = 4 * x * 6')
plt.grid()  
plt.xlabel('x axis')
plt.ylabel('y axis')
```

2. y = 10X^2 + 2X

```python
x = np.linspace(0,10,100)
y = np.sin(10*x**2+2*x)
plt.plot(x, y)
plt.title('y = 10 * x ^ 2 + 2 * x')
plt.grid()  
plt.xlabel('x axis')
plt.ylabel('y axis')
```
